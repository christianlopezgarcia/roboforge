```python
# Install dependencies if you haven’t already:
# pip install adafruit-circuitpython-pca9685

import time
import board
import busio
from adafruit_pca9685 import PCA9685
from adafruit_motor import servo

# Create I2C bus interface
i2c = busio.I2C(board.SCL, board.SDA)

# Create a PCA9685 object
pca = PCA9685(i2c)
pca.frequency = 50  # Typical servo frequency (50Hz)

# Create a servo object on channel 15 (last pin)
my_servo = servo.Servo(pca.channels[15])

# Move the servo back and forth
print("Moving servo on channel 15...")
while True:
    my_servo.angle = 0    # Move to 0 degrees
    time.sleep(1)
    my_servo.angle = 90   # Move to 90 degrees
    time.sleep(1)
    my_servo.angle = 180  # Move to 180 degrees
    time.sleep(1)
```

```python
import time
from adafruit_servokit import ServoKit

# Full 16-channel PCA9685
kit = ServoKit(channels=16)

# Last pin = channel 15 (not 6)
myservo = 8
kit.servo[myservo].angle = 180
time.sleep(1)

# Example: run continuous servo briefly, the n stop
kit.continuous_servo[1].throttle = 1   # full speed forward
time.sleep(1)
kit.continuous_servo[1].throttle = -1  # full speed reverse
time.sleep(1)
kit.continuous_servo[1].throttle = 0   # stop (hold)
# or: kit.continuous_servo[1].throttle = None  # stop (release completely)

# Reset last pin servo
kit.servo[myservo].angle = 0
time.sleep(1)
```

```python
# xarm_test.py
from xarm.wrapper import XArmAPI
import time

# =========================
# 1. Connect to the robot
# =========================
# Replace with your robot's IP address
arm = XArmAPI('192.168.1.164')

# Clear any warnings or errors
arm.clean_warn()
arm.clean_error()

# =========================
# 2. Enable and ready the robot
# =========================
arm.motion_enable(True)
arm.set_mode(0)       # Position control mode
arm.set_state(0)      # Ready state

# =========================
# 3. Read current position
# =========================
code, position = arm.get_position(is_radian=False)
print("Current position:", position)

# =========================
# 4. Move slightly
# =========================
# Move +50mm in X direction (if safe)
new_x = position[0] + 50
new_pos = [new_x, position[1], position[2],
           position[3], position[4], position[5]]

print("Moving to:", new_pos)
arm.set_position(*new_pos, speed=50, wait=True)
time.sleep(1)

# =========================
# 5. Disconnect
# =========================
arm.set_mode(0)
arm.move_gohome()
arm.disconnect()
print("Done.")
```

```python
#!/usr/bin/env python3
# Software License Agreement (BSD License)
#
# Copyright (c) 2019, UFACTORY, Inc.
# All rights reserved.
#
# Author: Vinman <vinman.wen@ufactory.cc> <vinman.cub@gmail.com>

"""
Description: Move Arc line(linear arc motion)
"""

import os
import sys
import time

# sys.path.append(os.path.join(os.path.dirname(__file__), '../../..'))

from xarm.wrapper import XArmAPI


# #######################################################
# """
# Just for test example
# """
# if len(sys.argv) >= 2:
#     ip = sys.argv[1]
# else:
#     try:
#         from configparser import ConfigParser
#         parser = ConfigParser()
#         parser.read('../robot.conf')
#         ip = parser.get('xArm', 'ip')
#     except:
#         ip = input('Please input the xArm ip address:')
#         if not ip:
#             print('input error, exit')
#             sys.exit(1)
# ########################################################


arm = XArmAPI(ip)
arm.motion_enable(enable=True)
arm.set_mode(0)
arm.set_state(state=0)

arm.move_gohome(wait=True)

paths = [
    [300, 0, 150, -180, 0, 0],
    [300, 200, 250, -180, 0, 0],
    [500, 200, 150, -180, 0, 0],
    [500, -200, 250, -180, 0, 0],
    [300, -200, 150, -180, 0, 0],
    [300, 0, 250, -180, 0, 0],
    [300, 200, 350, -180, 0, 0],
    [500, 200, 250, -180, 0, 0],
    [500, -200, 350, -180, 0, 0],
    [300, -200, 250, -180, 0, 0],
    [300, 0, 350, -180, 0, 0],
]

arm.set_position(*paths[0], wait=True)
_, angles = arm.get_servo_angle()
arm.set_pause_time(0.2)


def move():
    ret = arm.set_servo_angle(angle=angles, speed=50, wait=False)
    if ret < 0:
        print('set_servo_angle, ret={}'.format(ret))
        return -1
    for path in paths:
        ret = arm.set_position(*path[:6], radius=0, wait=False, speed=300)
        if ret < 0:
            print('set_position, ret={}'.format(ret))
            return -1
    return 0

for i in range(10):
    if move() != 0:
        break

arm.move_gohome(wait=True)
arm.disconnect()
```
